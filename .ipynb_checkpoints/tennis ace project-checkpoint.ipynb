```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
import seaborn as sns
```

 #load and investigate the data here:

```python
df = pd.read_csv("tennis_stats.csv")
df
# use correlation to determine the association and find the most useful colmns
df_considered_postive = df.corr()[(df.corr()>0.8) & (df.corr()<1.0)]
df_considered_postive =df_considered_postive[["Wins","Losses","Winnings","Ranking"]]
df_considered_negative = df.corr()[(df.corr()< -0.8) & (df.corr()>-0.99)]
df_considered_negative  =df_considered_negative[["Wins","Losses","Winnings","Ranking"]]
df_considered_postive
```

```python
# build a single feature linear regression model on aces to predict wins
# seperate the samples
x= df[["Aces"]]
y= df.Winnings
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=2)
# create regression obj
regr =LinearRegression()
regr.fit(x_train,y_train)
y_intercept = regr.predict(x_test)
plt.scatter(y_test,y_intercept,alpha = 0.4)
plt.xlabel("actual wins")
plt.ylabel("predicted wins")
plt.show()
regr.score(x_test,y_test)
with the above score aces alone are not cannot predict wins
```
